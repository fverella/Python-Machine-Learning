```python
# description de img4 
img4.shape

#transformation de l'image en gray 
from numpy import linalg

#construire le tableau avec des valeurs comprises entre 0 et 1

img_array = img4 / 255
print(" dimension du nouveau tableau img_array: ", img_array.shape)
type(img_array)
# il est possible d'obtenir une version en niveaux de gris assez raisonnable de notre image couleur si l'on applique la formule
#Y = 0,2126 R + 0,7152 G + 0,0722 B + 0,02 X
# où Y est le tableau représentant l'image en niveaux de gris, et R, G ,B X eont les tableaux de canaux rouge, vert et bleu et un X 
# que nous avions à l'origine
img_gray = img_array @ [0.2126, 0.7152, 0.0722,0.02]
print (" nouveau format en gray de img_array: ", img_gray.shape)
#Impression de l'image en gray
plt.imshow(img_gray, cmap="gray")
#Décomposition en valeurs singulieres de la matrice M=img_gray
U, s, Vt = linalg.svd(img_gray)
print(" dimension des différente matrices de la décomposition SVD")
print (U.shape,s.shape,Vt.shape)

import numpy as np
Sigma = np.zeros((418, 475))
for i in range(418):
    Sigma[i, i] = s[i]

#calcul de la distance entre les deux matrices M et approximation de M
distance = linalg.norm(img_gray - U @ Sigma @ Vt)    
print (" L'erreur mesurée par la distance entre les deux matrices est: " , distance)
```
